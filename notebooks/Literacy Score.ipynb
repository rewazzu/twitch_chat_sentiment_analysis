```python
messages = df[df.Vod_id == 858025579].Message
```

```python
def strip_emotes(message, emotes):
    if emotes:
        new_s = ''
    else:
        new_s = message
    for emote in emotes:
        new_s += message[:emote['begin']]
        new_s += message[emote['end']+1:]
    return new_s
```

```python
df['message_without_emotes'] = df.apply(lambda row: strip_emotes(row.Message, row.all_emotes), axis=1)
```

```python
messages = df[~(df.message_without_emotes == '')].message_without_emotes
```

```python
#s = messages.str.cat(sep='. ')
s = '. '.join(messages)
```

```python
readability = textstat.text_standard(s)
```
